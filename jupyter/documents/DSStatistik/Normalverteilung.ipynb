```python
import math
```

```python
def normal_cdf(x, mu=0, sigma=1):
    'Cumulative distribution function for the normal distribution'
    return (1.0 + math.erf((x - mu) / math.sqrt(2.0) / sigma)) / 2.0
```

```python
phi = normal_cdf(1.96)
phi
```

```python
from matplotlib import pyplot as plt
```

```python
xs =[x / 10.0 for x in range(-50,50)]
plt.plot(xs,[normal_cdf(x, sigma=1) for x in xs], '-', label='PHI(x) i.e. mu=0,sigma=1')
plt.plot(xs,[normal_cdf(x, sigma=2) for x in xs], '--', label='mu=0,sigma=2')
plt.plot(xs,[normal_cdf(x, sigma=0.5) for x in xs], ':', label='mu=0,sigma=0.5')
plt.plot(xs,[normal_cdf(x, mu=-1) for x in xs], ':', label='mu=-1,sigma=0.5')
plt.legend(loc=4) # bottom right
plt.title("Various cumulative distribution functions (CDF) of the normal distribution")
plt.show
```
